```python
import re

def hangul(text):
    return re.sub('[^ㄱ-ㅎ|ㅏ-ㅣ|가-힣]', ' ', text)
```

```python
wb_copy = wb.copy()
wb_copy['story'] = wb_copy['story'].apply(lambda x:hangul(x))

wb_copy.head(3)
```

```python
from konlpy.tag import Okt
okt = Okt() # Okt : Open Korean Text

def okt_tokenizer(text):
    filtered_words = []
    malist = okt.pos(text, norm=True, stem=True)
    
    for word in malist:
        filtered_words.append(word[0])
    return filtered_words
```

```python
stories = ''
for story in wb_copy['story']:
    stories += story + ' '

print(stories[:30])
```

```python
print(len(stories))
```

```python
word_list = okt_tokenizer(stories)
```

```python
word_list[:10]
```

```python
word_dict = {}
for word in word_list:
    if word in word_dict:
        word_dict[word] += 1
    else:
        word_dict[word] = 1
```

```python
print(len(word_dict))
```

```python
i = 1
for word, cnt in sorted(word_dict.items(), key=lambda x:x[1], reverse=True):
    print(f'{word:3} : {cnt:3}', end='  \t')
    if i % 10 == 0:
        print()
    i += 1
```

```python
word_df = pd.DataFrame(data={'word':word_dict.keys(), 'frequency':word_dict.values()})
```

```python
word_df
```

```python
word_df = word_df.sort_values(by='frequency', ascending=False)
```
